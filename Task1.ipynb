Now we will use sklearn to build a matrix in which the horizontal axis will have the given words as parameters and the vertical axis will have the alphabets

```python
def char_tokenizer(text):  # Define a tokenizer function to split each word into letters
    return list(text.lower())

data_prof = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/TaskData.xlsx',header = None) # Read the Excel file using pandas library

profesion = data_prof.iloc[:, 0].tolist() # Convert the data to a list of documents

vectorizer = CountVectorizer(analyzer='char', tokenizer=char_tokenizer) # Initialize CountVectorizer with tokenizer

X = vectorizer.fit_transform(profesion) # Fit and transform the documents

feature_names = vectorizer.get_feature_names_out() # Get feature names (characters)

char_to_index = {char: i for i, char in enumerate(sorted(set(feature_names)))} # Create a dictionary to map characters to indices

complete_alphabet = list(string.ascii_lowercase) # Create a complete alphabet list

# Create a new matrix with zeros for missing characters
matrix_1 = []
for row in X.toarray():
    new_row = [0] * len(complete_alphabet)
    for i, char in enumerate(complete_alphabet):
        if char in char_to_index:
            new_row[i] = row[char_to_index[char]]
    matrix_1.append(new_row)

print(matrix_1)
print(len(matrix_1))
```

Now we will write a matrix for the professtionals

```python

def char_tokenizer(word):
  return list(word.lower())


data_prof_1 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/prof.xlsx',header = None) # Read the Excel file

# Convert the data to a list of documents
profesion_1 = data_prof_1.iloc[:, 0].tolist()

# Initialize CountVectorizer with custom tokenizer
vectorizer_1 = CountVectorizer(analyzer='char', tokenizer=char_tokenizer)

# Fit and transform the documents
X_1 = vectorizer_1.fit_transform(profesion_1)

# Get feature names (characters)
feature_names = vectorizer_1.get_feature_names_out()

# Create a dictionary to map characters to indices
char_to_index = {char: i for i, char in enumerate(sorted(set(feature_names)))}

# Create a complete alphabet list
complete_alphabet = list(string.ascii_lowercase)

# Create a new matrix with zeros for missing characters
new_matrix1 = []
for row in X_1.toarray():
    new_row = [0] * len(complete_alphabet)
    for i, char in enumerate(complete_alphabet):
        if char in char_to_index:
            new_row[i] = row[char_to_index[char]]
    new_matrix1.append(new_row)

print(new_matrix1)
print(len(new_matrix1))
```

Now we will apply cosine similarity and then therough few steps finally plot the bar graph

```python
def cos_sim(a,b):




    dot_product = np.dot(a, b)# Compute the dot product between x and y


    magnitude_a = np.sqrt(np.sum(a**2)) # calculate the magnitude of vector a and b
    magnitude_b = np.sqrt(np.sum(b**2))


    if magnitude_a == 0 or magnitude_b == 0:    # Check for zero vectors
        return 0


    cosine_similarity = dot_product / (magnitude_a * magnitude_b) # calculate the cosine similarity

    return cosine_similarity
```

Now we will decide which word corresponds to which profession and save them in a numpy array

```python
array_lamba = np.zeros(10, dtype=np.float64)

for i in range(len(matrix_1)):
  max_number = float(0)
  for j in range(len(new_matrix1)):
    number = cos_sim(np.array(matrix_1[i]),np.array(new_matrix1[j]))
    if max_number < number:
      max_number = number
      index = j
  array_lamba[index] += 1

print(array_lamba)
```

```python
import matplotlib.pyplot as plt

# Example array of integers

data = array_lamba
# Create a list of positions for the bars on the x-axis
positions = range(len(data))
x_labels = ['Doctor', 'Lawyer', 'Teacher', 'Engineer', 'Accountant', 'Nurse', 'Police', 'Architect', 'Dentist', 'Pharmacist']
# Create the bar graph
plt.bar(positions, data, color='blue')


# Add labels to each bar
for i, height in enumerate(data):
    plt.text(i, height + 0.5, str(height), ha='center')
# Set the x-axis labels
plt.xticks(positions, x_labels,rotation = 'vertical')

# Add labels and title
plt.xlabel('Professions')
plt.ylabel('frequency')
plt.title('TASK_1')

# Show the plot
plt.show()
```
